```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
books = pd.read_csv('E:\\Projects\\Book Recommender\\Dataset\\books.csv')
```

```python
books.head()
print(len(books))
```

```python
ax=plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")

plt.show()
```

```python
import numpy as np
```

```python
books['missing_desc'] = np.where(books['description'].isna(),1,0)
books['age_of_book']=2025-books['published_year']
```

```python
columns_of_interest = ["num_pages","age_of_book","missing_desc","average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style="white")

plt.figure(figsize=(8,6))
heatmap=sns.heatmap(correlation_matrix,annot=True,fmt=".2f",cmap="coolwarm", cbar_kws={"label":"Spearman correlation"})

plt.show()
```

```python
# to maintain rows where the values are present instead of NaN
book_missing = books[~(books["description"].isna()) &
    ~(books["num_pages"].isna()) &
    ~(books["average_rating"].isna()) &
    ~(books["published_year"].isna())
    ]
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
book_missing["words_in_desc"] = book_missing["description"].str.split().str.len()
```

```python
book_missing_25_words = book_missing[book_missing["words_in_desc"]>=25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(),book_missing_25_words["title"],
            book_missing_25_words[["title","subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_desc"] = book_missing_25_words[["isbn13","description"]].astype(str).agg(" ".join,axis=1)
```

```python
(
    book_missing_25_words.drop(["subtitle","missing_desc","age_of_book","words_in_desc"],axis=1)
    .to_csv("books_cleaned.csv",index=False)
)
```
